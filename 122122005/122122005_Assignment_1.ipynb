```python
from sklearn.metrics import accuracy_score
class MPNeuron:
    def __init__(self):
        self.b = None
        
    def model(self, x):
        return(np.sum(x)>=self.b)
    
    def predict(self, X):
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)
    
    def fit(self, X, Y):
        accuracy = {} 
        for b in range(X.shape[1]+1):
            self.b = b
            y_pred_train = self.predict(X)
            accuracy[b] = accuracy_score(y_pred_train, Y)
        
        best_b = max(accuracy, key = accuracy.get)
        self.b = best_b
        
        print("optimal value of b: ", self.b)
        print("Accuracy for train data: ", accuracy[self.b])
        
    def test(self, X, Y):
        y_pred_train = self.predict(X)
        accuracy = accuracy_score(y_pred_train, Y)
        print("Accuracy for test data: ", accuracy)
            
```

```python
mpneuron = MPNeuron()
mpneuron.fit(x_binarised_train, y_train)
```

```python
mpneuron.test(x_binarised_test, y_test)
```

## Perceptron Model

```python
class Perceptron:
    
    def __init__(self, learning_rate=0.01, iterations = 1000):
        self.bias = None
        self.weights = None
        self.activation_func = self.unit_step_func
        self.lr = learning_rate
        self.epoch = iterations
        
    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_pred = self.activation_func(linear_output)
        return y_pred
        
    def unit_step_func(self, x):
        return np.where(x>=0,1,0)
    
    def fit(self, X , y):
        #X is MxN    M rows/samples and N features/columns
        n_samples, n_features = X.shape
        
        #initialize weights to each feature as zero
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        #ensure that y is 0 or 1
        y_ = np.array([1 if i>0 else 0 for i in y])
        
        
        #percepron model will converge in finite iterations
        for epoch in range(self.epoch):
            for p,q in enumerate(X):
                linear_output = np.dot(q, self.weights) + self.bias
                y_pred = self.activation_func(linear_output)
                
                #update the weights and bias as per the formula
                # W = W + (lr * (y - y_pred) * x)
                # bias = bias + (lr * (y - y_pred))
                
                update = self.lr * (y_[p] - y_pred)
                self.weights += update * q
                self.bias += update
```

```python
#x_train, x_test, y_train, y_test
#convert x_train, x_test into arrays
x_train = x_train.values
print(type(x_train))
x_test = x_test.values
print(type(x_test))
```

```python
def accuracy(y_test, y_pred):
    accuracy = np.sum(y_test == y_pred) / len(y_test)
    return accuracy
```

```python
p = Perceptron(learning_rate = 0.1, iterations=1000)
p.fit(x_train,y_train)
```

```python
prediction = p.predict(x_test)
print("Perceptron Classification accuracy is: ",accuracy(y_test, prediction))
```
